```python
# Import the necessary packages
import numpy as np
import pandas as pd

import statsmodels.api as sm
from statsmodels.tsa.stattools import coint
from statsmodels.regression.rolling import RollingOLS
from statsmodels.regression.linear_model import OLS
from statsmodels.tsa.tsatools import lagmat, add_trend
from statsmodels.tsa.adfvalues import mackinnonp

import yfinance as yf
import seaborn
import matplotlib.pyplot as plt
```

```python
# Select the stocks we would like to consider, in this case we look at US airlines
tickers = ['DAL', 'LUV', 'UAL', 'AAL', 'ALK', 'SKYW']
prices = yf.download(tickers, start='2023-01-01', end='2023-12-31')['Adj Close']
prices
```

```python
# Create a function that checks for cointegration
def find_cointegrated_pairs(data):
  n = data.shape[1]
  score_matrix = np.zeros((n, n))
  pvalue_matrix = np.ones((n, n))
  keys = data.keys()
  pairs = []
  for i in range(n):
    for j in range(i+1, n):
      S1 = data[keys[i]]
      S2 = data[keys[j]]
      result = coint(S1, S2)
      score = result[0]
      pvalue = result[1]
      score_matrix[i, j] = score
      pvalue_matrix[i, j] = pvalue
      if pvalue < 0.05:
        pairs.append((keys[i], keys[j]))
  return score_matrix, pvalue_matrix, pairs
```

```python
# Apply this to the data
scores, pvalues, pairs = find_cointegrated_pairs(prices)
```

```python
# Plot to see which stocks are cointegrated
seaborn.heatmap(pvalues, xticklabels = tickers, yticklabels= tickers, cmap = 'RdYlGn_r',mask = (pvalues >= 0.05))
```

```python
# Conduct ADF unit root test
def adf(ts, maxlag=1):
  # Convert to array, determine dimension and calculate discrete difference
  ts = np.asarray(ts)
  dim = ts.shape[0]
  tsdiff = np.diff(ts)

  # Generate 2d array of lags and remove invalid observations
  tsdall = lagmat(tsdiff[:,None], maxlag, trim='both', original='in')
  # Determine dimension
  nobs = tsdall.shape[0]
  # Substitute 0xdiff with level of x
  tsdall[:,0] = ts[-nobs - 1:-1]
  tsdshort = tsdiff[-nobs:]

  # Determine the linear regression with OLS
  results = OLS(tsdshort, add_trend(tsdall[:, :maxlag+1],'c')).fit()
  adfstat = results.tvalues[0]
  # Get appoximated p value
  pvalue = mackinnonp(adfstat, 'c', N=1)
  return pvalues

def cadf(a, b):
  # Get cointegrated Augmented Dickey Fuller Test
  ols = OLS(a,b).fit()
  return adf(ols.resid)
```

```python
# Use ADF to check for cointegration for DAL and AAL
S1 = prices.DAL
S2 = prices.AAL

cadf(S1, S2)
```

```python
# Determine the hedge ratio
S1 = sm.add_constant(S1)
results = sm.OLS(S2, S1).fit()
S1 = S1.DAL
b = results.params['DAL']
spread = S2 - b*S1
```

```python
# Plot it
spread.plot()
plt.axhline(spread.mean(), color='red')
plt.legend(['Spread'])
```

```python
# Determine the Z score and plot it
def zscore(data):
  return (data - data.mean())/np.std(data)

zscore(spread).plot()
plt.axhline(zscore(spread).mean(), color = 'black')
plt.axhline(1.0, color='red', linestyle = '--')
plt.axhline(-1.0, color='green', linestyle = '--')
plt.legend(['Spread z-score', 'Mean', '+1', '-1'])
```

```python
# Create a DataFrame with the signal and position size in the paid
trades = pd.concat([zscore(spread), S2-b*S1], axis=1)
trades.columns = ["signal","position"]

# Add a long and short position at the z score levels
trades["side"] = 0.0
trades.loc[trades.signal <= -1, "side"] = 1
trades.loc[trades.signal >= 1, "side"] = -1
```

```python
returns = trades.position.pct_change()*trades.side
returns.cumsum().plot()
```
